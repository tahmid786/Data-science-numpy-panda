```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
a = np.arange(11)
b = a**2
plt.show()

plt.scatter(a, b)

plt.plot(a,b,'r*')
plt.show()
```

```python
plt.plot(a,b,'r*')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("figure")
plt.show()
```

```python
plt.subplot(1,3,1)
plt.plot(a,b,'g--')
plt.subplot(1,3,2)
plt.plot(b,a,'r*')
plt.subplot(1,3,3)
plt.plot(a,b,'bo-')
plt.show()
```

# object oriented approach

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
a = np.arange(11)
b = a**2
fig1=plt.figure(figsize=(6,7))
axes=fig1.add_axes([.3,.5,.8,.5])
axes.plot(a,b,'r--*')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('FIGURE')
fig1.savefig("a.jpeg",dpi=200)
```

```python
fig,axes=plt.subplots(2,2)
plt.tight_layout()
```

```python
fig,axes=plt.subplots(2,2,sharex=True,sharey=True,figsize=(7,8))
plt.tight_layout()
axes[0][0].plot(a,b,'bo-')
axes[1][1].plot(a,b,'r*-')
axes[1][0].plot(a,b,'g-')
axes[0][1].plot(a,b,'y--')
```

```python
fig2=plt.figure(figsize=(10,5))
axes_main=fig2.add_axes([.1,.1,.8,.6])
axes_in=fig2.add_axes([.3,.4,.2,.2])
axes_main.plot(a,b,'y--')
axes_main.set_xlabel('main x')
axes_main.set_ylabel('main y')
axes_in.plot(b,a,'b+')
axes_in.set_xlabel('in x')
axes_in.set_xlabel('in y')
axes_main.set_title('Graph')
fig2.savefig('D:\\figure.pdf',dpi=100)
```

```python
fig3=plt.figure()
axes=fig3.add_axes([0,0,1,1])
axes.plot(a,a**2,label="power2")
axes.plot(a,a**3,label="power3")
axes.plot(a,a**4-a**3,label="power4")
axes.legend()
```
